```python
#writing csv from dataframe and saving it
def write_csv_from_df(df, filename):
    df.to_csv(filename, index=False)    
```

```python
def get_player_tournament_data(year):
    if year == 2022:
        year = 2023
    file_name_denti = f"atp_senza_ripetizioni_{year}.csv"
    file_senza_denti = open(file_name_denti, "r", encoding="utf-8")
    lista_codici_players=[]

    for line in file_senza_denti:
        if line.strip() != "" and not line.startswith("Player Id"):
            lista_codici_players.append(line.strip().split(",")[0])
    print(lista_codici_players)
    print(len(lista_codici_players))
    return lista_codici_players
```

```python
from datetime import datetime
url_incomplete = f"https://www.atptour.com/en/-/www/activity/sgl/"
anni = [2022,2023,2024,2025]
columns=[
                "Year","EventId","PlayerCode","Date","EventName","EventType","EventCountry",
                "PlayerRank","PointsGained","PrizeUsd","SglDrawSize","Surface","TotPrizeMoney"
            ]

for y in anni:
    i=0
    df = pd.DataFrame(columns=columns)
    lista_codici_players = get_player_tournament_data(y)
    for c in lista_codici_players:
        i+=1
        print(f"Processing player {i}/{len(lista_codici_players)}: {c}")
        #per ongi giocatore estraggo tutti i tornei, uno per anno    
        url = f"https://www.atptour.com/en/-/www/activity/sgl/{c}/"
        print(url)
        response = requests.get(url, params=params, cookies=cookies, headers=headers)
        
        js = response.json()
        
        for yearly_activity in js["Activity"]:
            if yearly_activity["EventYear"] in anni:
                yearly_tournaments = yearly_activity["Tournaments"]
                year = yearly_activity["EventYear"]
                
                #creo un dataframe per ogni anno
                #df_filename = f"PROVA_data_{year}.csv"
                
                
                #per ogni anno ora prendo ogni singolo torneo, TUTTI I TIPI
                for t in yearly_tournaments:
                    if t["Display"]==True: #there are some "strange" tournaments with Display=False
        
                        player_code = c
                        event_id = t["EventId"]
                        #data di fine torneo
                        event_date = datetime.fromisoformat(t["EventDate"]).date()
                        event_name = t["EventName"]
                        event_type = t["EventType"]
                        event_location = t["Location"]["EventLocation"]
                        player_rank = t["PlayerRank"]
                        points_gained = t["Points"]
                        player_prize = t["PrizeUsd"]
                        event_draw_size = t["SglDrawSize"]
                        event_surface = t["Surface"]
                        event_tot_prize_money = t["TotPrizeMoney"]
                        
                        row = {
                            "EventId": event_id,
                            "Date": event_date,
                            "EventName": event_name,
                            "EventType": event_type,
                            "EventCountry": event_location,
                            "PlayerRank": player_rank,
                            "PointsGained": points_gained,
                            "PrizeUsd": player_prize,
                            "SglDrawSize": event_draw_size,
                            "Surface": event_surface,
                            "TotPrizeMoney": event_tot_prize_money,
                            "Year": year,
                            "PlayerCode": player_code}
                        
                        if year == y:
                            df.loc[len(df)]=row
                            
                        #yearly_df = yearly_df.loc[len(yearly_df)]=row
                                        
                        # if i==10:
                        #     break
    write_csv_from_df(df, f"df_{y}.csv")
    print(f"Saved file: {df}")
    
```
