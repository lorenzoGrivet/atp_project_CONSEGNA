```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("analisi_pts_difendere/25_pts_24.csv")
df["perc"] = df["perc_tornei_con_punti_da_difendere"] * 100
df = df.sort_values("perc")

fig, ax = plt.subplots(figsize=(6, 3.5))
bars = ax.bar(df["gruppo"], df["perc"])
ax.set_ylabel("% tornei con punti da difendere")
ax.set_ylim(0, 50)

ax.bar_label(bars, fmt="%.1f%%", padding=3)
plt.show()
```

```python
import pandas as pd

df23 = pd.read_csv("analisi_pts_difendere/23_pts_22.csv")
df24 = pd.read_csv("analisi_pts_difendere/24_pts_23.csv")
df25 = pd.read_csv("analisi_pts_difendere/25_pts_24.csv")

df23["year"] = 2023
df24["year"] = 2024
df25["year"] = 2025

df = pd.concat([df23, df24, df25], ignore_index=True)


import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(7, 4))
sns.barplot(
    data=df,
    x="gruppo",
    y="perc_tornei_con_punti_da_difendere",
    hue="year"
)
plt.ylim(0, 1)
plt.ylabel("Perc. tornei con punti da difendere")
plt.xlabel("Gruppo ranking")
plt.title("Punti da difendere per gruppo e anno")
plt.legend(title="Anno")
plt.show()
```

```python
plt.figure(figsize=(7, 4))
sns.lineplot(
    data=df,
    x="year",
    y="perc_tornei_con_punti_da_difendere",
    hue="gruppo",
    marker="o"
)
plt.ylim(0, 1)
plt.ylabel("Perc. tornei con punti da difendere")
plt.xlabel("Anno")
plt.title("Trend (20232025)")
plt.show()
```
