```python

def get_tournaments2024_25(response):

    soup = bs(response.text)
    table = soup.findAll('table',class_="wikitable")

    list = []
    for t in table:
        ll = t.findAll('tr',style="vertical-align:top")

        for l in ll:
            cells = l.findAll('td')
            for c in cells[:2]:
                text_cell = c.get_text(strip=True)

                if text_cell.split(" ")[1] in ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]:
                    date = text_cell

                if ("ATP 250" in text_cell) or ("ATP 500" in text_cell):
                    location = text_cell

                    list.append([date, location])
    return list
```

```python
def get_tournaments2023(response):

    soup = bs(response.text)
    table = soup.findAll('table',class_="wikitable")

    list = []
    for t in table:
        ll = t.findAll('tr',valign="top")

        for l in ll:
            cells = l.findAll('td')
            for c in cells[:2]:
                text_cell = c.get_text(strip=True)

                if text_cell.split(" ")[1] in ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]:
                    date = text_cell

                if ("ATP 250" in text_cell) or ("ATP 500" in text_cell):
                    location = text_cell
                
                    list.append([date, location])
    return list
    
```

```python
import pandas as pd
import csv

def parse_row_no_regex(text):
    # Category
    category = "ATP 500" if "ATP 500" in text else "ATP 250"

    # Surface
    if "Hard" in text:
        surface = "Hard"
    elif "Clay" in text:
        surface = "Clay"
    elif "Grass" in text:
        surface = "Grass"
    else:
        surface = ""

    # Prize money
    prize_money = ""
    for part in text.split("–"):
        part = part.strip()
        if part.startswith("$") or part.startswith("€"):
            prize_money = part[1:]
            break

    # fisrt part of category
    before_cat = text.split(category)[0]

    # Country and tournament
    if "," in before_cat:
        country = before_cat.split(",")[-1].strip()
        tournament = before_cat.split(",")[0].strip()
    else:
        country = "Hong Kong SAR"
        tournament = country


    return pd.Series([tournament, country, category, surface, prize_money])

dfs = []
for year in range(2023,2026):
    file = f"atp_tournaments_{year}.csv"
    df = pd.read_csv(file)
    df[["Tournament", "Country", "Category", "Surface", "PrizeMoney"]] = (
        df["Location"].apply(parse_row_no_regex)
    )
    df["Date"] = pd.to_datetime(df["Date"] + f" { year}", format="%d %b %Y")

    df = df[["Date", "Tournament", "Country", "Category", "Surface", "PrizeMoney"]]
    dfs.append(df)

# Final Dataframe
final_df = pd.concat(dfs, ignore_index=True)

# Sort the Dataframe
final_df = final_df.sort_values("Date")

output = "tournaments_clean_all_years.csv"
final_df.to_csv(output, index=False)
final_df
```
