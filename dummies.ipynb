```python
import pandas as pd

years = [23, 24, 25]

for year in years:
    df = pd.read_csv(f"df_aggiunti_pts_def/df_{year}.csv")

    # drop_first=True to avoid collinearity during regression
    df = pd.get_dummies(df, columns=["EventType", "Surface"], prefix=["EventType", "Surface"], dtype=int, drop_first=True)

    out = df.copy()
    out.to_csv(f"df_con_dummies/df_{year}.csv")
```
